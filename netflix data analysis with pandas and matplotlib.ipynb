```python
df['director']=df['director'].fillna("Unknown")
df['cast']=df['cast'].fillna("Unknown")
df['country']=df['country'].fillna("Unknown")
df['rating']=df['rating'].fillna("Not Rated")
df.dropna(subset='date_added',inplace=True)
```

```python
df.dropna(subset='duration',inplace=True)
```

```python
len(df) #will count the total num of rows excluding header
```

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce') # the coerce will ensure if value isnt a valid date it becomes NAT (missing)
```

```python
df.dropna(subset='date_added',inplace=True)
df.info()
```

```python
df.to_csv("D:\\netflix_titles.csv",index=False)
df
```

```python
df_genres = df.copy()
df_genres['listed_in'] = df_genres['listed_in'].str.split(', ')
df_genres = df_genres.explode('listed_in')

top_genres = df_genres['listed_in'].value_counts().head(10)
df_genres
top_genres # most common genre in netflix
```

```python
df_genres['listed_in'].value_counts().head(10).plot(kind='bar',color='skyblue')
```

```python
df['type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Content type count')
```

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='bar',color='orange')
plt.title('Content added over time on netflix')
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line',marker='o')
plt.title('Release Year Distribution')
```

```python
df_filtered = df[df['country'] != 'Unknown']  #to remove the unknown row so it don't get counted.

df_filtered['country'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Countries by Content (Excluding Unknown)')
plt.xlabel('Country')
plt.ylabel('Number of Contents')
plt.show()
```

```python
# Step 1: Get top 10 genres list (not value_counts)
top_10_genre_list = df_genres['listed_in'].value_counts().head(10).index.tolist()

# Step 2: Crosstab and plot
pd.crosstab(df_genres['listed_in'], df_genres['type']) \
.loc[top_10_genre_list] \
.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set3')

plt.title("Top 10 Genres ")
```

```python
# Step 1: Drop nulls or unknowns
df_directors = df[df['director'].notnull() & (df['director'].str.strip().str.lower() != 'unknown')]

# Step 2: Get top 10 directors by frequency
top_10_directors = df_directors['director'].value_counts().head(10)

# Optional: Plot
top_10_directors.plot(kind='barh', color='purple',figsize=(8,5))

plt.title("Top 10 Directors")
```

```python

import seaborn as sns

# Group by year and type
content_trend = df.groupby(['release_year', 'type']).size().reset_index(name='count')

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=content_trend, x='release_year', y='count', hue='type', marker='o')
plt.title('Netflix Content Added Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Drop nulls in 'cast'
df_cast = df.dropna(subset=['cast'])

# Split multiple actors and explode into rows
all_actors = df_cast['cast'].str.split(', ').explode()

# Remove 'Unknown' entries
all_actors = all_actors[all_actors.str.lower() != 'unknown']

# Get top 10 most frequent actors
top_actors = all_actors.value_counts().head(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=top_actors.values, y=top_actors.index, palette='magma')
plt.title('Top 10 Most Frequent Actors/Actresses on Netflix (Excluding Unknown)')
plt.xlabel('Number of Appearances')
plt.ylabel('Actor/Actress')
plt.tight_layout()
plt.show()
```
